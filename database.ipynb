```python
db = sql.connect('./data/NBA.db')
dbCursor = db.cursor()
```

```python
# 創建 teams 表格
dbCursor.execute('''
CREATE TABLE IF NOT EXISTS teams (
    teamid INTEGER PRIMARY KEY,
    name TEXT NOT NULL,
    abbreviation TEXT NOT NULL UNIQUE
)
''')

# 創建 players 表格
dbCursor.execute('''
CREATE TABLE IF NOT EXISTS players (
    playerid INTEGER PRIMARY KEY,
    name TEXT NOT NULL,
    team TEXT,
    FOREIGN KEY (team) REFERENCES teams(abbreviations)
)
''')
db.commit()
```
