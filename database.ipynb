```python
db = sql.connect('./data/NBA.db')
dbCursor = db.cursor()
```

```python
dbCursor.execute('''
CREATE TABLE IF NOT EXISTS players (
    playerid INTEGER PRIMARY KEY,
    name TEXT NOT NULL
)
''')

dbCursor.execute('''
CREATE TABLE IF NOT EXISTS teams (
    teamid INTEGER PRIMARY KEY,
    name TEXT NOT NULL
)
''')
db.commit()
```

```python
for playerid, name in zip(unique_playerids, unique_names):
    dbCursor.execute('INSERT INTO players (playerid, name) VALUES (?, ?)',  (int(playerid), str(name)))
db.commit()
```

```python
for teamid, name in team_abbreviations.items():
    dbCursor.execute('''
    INSERT INTO teams (teamid, name) VALUES (?, ?)
    ''', (teamid, name))
db.commit()
```

```python
db.commit()
db.close()
```
